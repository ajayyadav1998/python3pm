```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.linspace(.1,.5,11)
```

```python
x
```

```python
y = x**2
y
```

```python
plt.plot(x,y)
plt.title('plot of x and y')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

```python
#subplot

plt.subplot(1,2,1)
plt.plot(x,y,'r')
plt.title('x over y')

plt.subplot(1,2,2)
plt.plot(y,x,'m')
plt.title('y over x')

plt.tight_layout()
```

```python
plt.subplots(nrows=2,ncols=2)
```

```python
# creation of graph with axes

fig = plt.figure(figsize=(10,6),facecolor='red')

ax1 = fig.add_axes([0,0,.8,.8]) #left,bottom,width,height
ax2 = fig.add_axes([.2,.4,.25,.25])


ax1.plot(x,y,'r:')
ax1.set_title('large graph')
ax1.set_xlabel('x-axis')
ax1.set_ylabel('y-axis')

ax2.plot(y,x,'y')
ax2.set_title('small graph')
ax2.set_xlabel('x-axis')
ax2.set_ylabel('y-axis')

plt.show()
```

```python
random = np.random.normal(50,20,1000)

plt.hist(random,bins=30,edgecolor='black')
plt.title('histogram')
plt.xticks(rotation=90)
plt.show()
```
