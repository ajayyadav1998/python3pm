```python
newdf = finaldf.copy()
```

```python
newdf['Survived'].fillna(0,inplace=True)
```

```python
newdf['Embarked'].fillna('S',inplace=True)
newdf['Pclass'].fillna(1,inplace=True)
newdf['Age'].fillna(newdf.Age.mean(),inplace=True)
newdf['Fare'].fillna(newdf.Fare.mean(),inplace=True)
```

```python
ndf = newdf.copy()
```

```python
newdf['Age']= newdf['Age'].apply(lambda x : int(x))
```

```python
newdf
```

# Number of adults and childrens in df

```python
newdf['Agestat'] = np.where(newdf['Age'] > 18,'adult','child')
```

```python
newdf['Agestat'].value_counts().plot.bar(title='children vs adults',figsize=(8,4),color=['pink','yellow'])
plt.show()
```

```python
age = newdf.Age
mean = newdf.Age.mean()
```

```python
plt.figure(figsize=(9,5))
plt.hist(age,bins=30,edgecolor='black',color='r',label='Age')
plt.title('Avg Age of Peoples in df')
plt.plot([mean,mean],[0,370],'g',lw=4)
plt.legend()
plt.plot()
#plt.show()
plt.savefig('C:/Users/Ajay/Desktop/my.png')
```
