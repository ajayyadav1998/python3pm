```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
sns.get_dataset_names()
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.shape
```

```python
tips.head()
```

```python
tips.sex.value_counts().plot(kind='bar')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(x='sex',data=tips,palette='cubehelix')
plt.title('males and females counting')
plt.show()
```

```python
tips.head()
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x='sex',y='size',data=tips,palette='tab20b_r')
plt.title('number of party members')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(x='sex',y='tip',data=tips,palette='rainbow')
plt.title('tips giving by males and females')
plt.show()
```

```python
tips.head()
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(x='sex',y='total_bill',data=tips,palette='hls')
plt.title('bill payable on the basis of sex')
plt.show()
```

```python
sns.palplot(sns.color_palette('Greens'),size=10)
sns.palplot(sns.color_palette('cubehelix'),size=10)
sns.palplot(sns.color_palette('Set1'),size=10)
sns.palplot(sns.color_palette('Set2'),size=10)
sns.palplot(sns.color_palette('Set3'),size=10)
sns.palplot(sns.color_palette('Dark2'),size=10)
sns.palplot(sns.color_palette('flag'),size=10)
sns.palplot(sns.color_palette('flag_r'),size=10)
sns.palplot(sns.color_palette('RdPu'),size=10)
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(data=tips,palette='RdPu')
plt.title('overall data box plot')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(x='sex',y='tip',hue='smoker',data=tips,palette='RdPu')
plt.title('tip giving by male and female on the basis of non-s/s')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.violinplot(x='sex',y='tip',data=tips,palette='PuBu',vert=False)
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.stripplot(x='sex',y='tip',data=tips,palette='Dark2',jitter=True)
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.swarmplot(x='sex',y='total_bill',data=tips)
plt.show()
```
