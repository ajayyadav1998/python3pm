```python
x = np.linspace(0,5,11)
y = x ** 3
```

```python
fig,ax = plt.subplots(nrows=2,ncols=2,figsize=(10,5))

plt.style.use('ggplot')

ax[0,0].plot([1,2,3,4,5],[1,4,9,16,25],'r--')
ax[0,1].plot([9,8,7,6,5],[81,64,49,36,25],'g-o')
ax[1,0].plot(x,y,'m:')
ax[1,1].plot(y,x,'y--')

plt.suptitle('Subplots')
plt.show()
```
