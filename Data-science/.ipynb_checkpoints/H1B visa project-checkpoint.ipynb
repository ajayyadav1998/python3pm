```python
# top 10 desirable states by employee
df.head()
```

```python
worksite = df['worksite'].value_counts().nlargest(10)
deny  = df[df['casestatus']=='DENIED']['worksite'].value_counts()[:5]

fig = plt.figure(figsize=(12,8),facecolor='yellow')

ax1 = fig.add_axes([0,0,.4,.4])
ax2 = fig.add_axes([.5,0,.4,.4])

ax1.bar(x=worksite.index,height=worksite[:])
ax1.set_xticklabels(labels=worksite.index,rotation=90)
ax1.set_title('top 10 desirable worksites')
ax1.set_xlabel('states names')
ax1.set_ylabel('numbers')

ax2.bar(x=deny.index,height=deny[:])
ax2.set_xticklabels(labels=deny.index,rotation=90)
ax2.set_title('top 5 denied worksites')
ax2.set_xlabel('states name')
ax2.set_ylabel('numbers')

plt.show()
```

```python
jobs =['DATA SCIENTIST','DATA ANALYST','DATA ENGINEER','BUSINESS ANALYST']
total = []

for var in jobs:
    w = df[df['jobtitle']==var]['jobtitle'].count()
    total.append(w)
    
    
plt.bar(x=jobs,height=total[:])
plt.xticks(rotation=90)
plt.xlabel('JOB TITLE')
plt.ylabel('NO. OF JOBS')
plt.title('companies which gives datascience post')
plt.show()
```
